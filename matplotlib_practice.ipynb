```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#skill drill
ax = plt.axes()
ax.plot(x_axis,y_axis,
        color="green",
        marker="D",
        linewidth=2,
        label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0,45)
ax.grid()
```

```python
ax = plt.axes()
ax.scatter(y_axis, x_axis,
           c="SkyBlue",
           label="Boston",
           alpha = .8,
           edgecolor = "Black",
           linewidth = 2,
           s=[i*5 for i in y_axis])
ax.legend()
ax.set_xlim(0,50)
ax.set_title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
#create pie chart
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#skill drill
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

ax = plt.axes()
ax.pie(y_axis,
       labels=x_axis,
       autopct='%.1f%%',
       explode=explode_values,
       shadow=True,
       startangle=95,
       counterclock=False,
       colors=colors)
plt.show()
```

```python
import statistics
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#skill drill
ax = plt.axes()
ax.errorbar(x_axis,y_axis,
        color="green",
        marker="D",
        linewidth=2,
        label="Boston",
        yerr=stdev,
        capsize=3)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")
ax.grid()
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
```

```python
ax = plt.axes()
ax.barh(x_axis,y_axis,
       color="cyan",
       label="Chicago",
       xerr=stdev,
       capsize=3)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_xticks(np.arange(0,65,step=5.0))
plt.gca().invert_yaxis()

ax.xaxis.set_minor_locator(MultipleLocator(1))
```
